```python
 
def draw_square(side): 
  for i in range(0,4): 
    forward(side) 
    right(90) 
 
def draw_board(cell_side):
  bgcolor("blue")
  color("white")
  for i in range (0,window_width,cell_side):
    for j in range (0,window_height+cell_side,cell_side):
      penup()
      goto(i,j)
      pendown()
      draw_square(cell_side)
 
def place_piece(col,row,piece_color,cell_side): 
  penup()
  goto(col*cell_side+25,int((5-row)*cell_side + cell_side/2))
  pendown()
  color(piece_color)
  width(10)
  
  for i in range(0,18): 
    forward(10) 
    right(20)
 
 
def ganador(grid, pieza):
    #checkeamos la posibilidad de una victoria en vertical
    for i in range(3):
        for j in range(7):
            if grid[i][j] == pieza and grid[i+1][j] == pieza and grid[i+2][j] == pieza and grid[i+3][j] == pieza:
                return True
    #checkeamos la posibilidad de una victoria en horizontal
    for i in range(6):
        for j in range(4):
            if grid[i][j] == pieza and grid[i][j+1] == pieza and grid[i][j+2] == pieza and grid[i][j+3] == pieza:
                return True
    #checkeamos la posibilidad de una victoria en diagonal (creciente)
    for i in range(3):
        for j in range(4):
            if grid[i][j] == pieza and grid[i+1][j+1] == pieza and grid[i+2][j+2] == pieza and grid[i+3][j+3] == pieza:
                return True
    #checkeamos la posibilidad de una victoria en diagonal(decreciente)
    for i in range(3, 6):
        for j in range(4):
            if grid[i][j] == pieza and grid[i-1][j+1] == pieza and grid[i-2][j+2] ==pieza and  grid[i-3][j+3] == pieza:
                return True
    return False
def contador_pieza1(grid, row, col):
    grid[row][col] = 1 #Sustituimos el valor en la posicion de la ficha en la matriz por 1

def contador_pieza2(grid, row, col):
    grid[row][col] = 2 #Sustituimos el valor en la posicion de la ficha en la matriz por 2

def fila_correcta(grid, col):
    row = 0 #Intenta colocar la ficha en la fila 0
    while grid[row][col] != 0: #Comprueba si esa posicion esta ocupada, si lo está, prueba por la fila siguiente, y asi hasta que encuentre sitio
        row = row+1
    return row #Nos devuelve el valor de la fila correcta

def columna_correcta(grid, col):
    while grid[5][col] != 0: #Comprueba si la columna esta llena
        col = int(input('Esa columna está llena, seleccione otra(La 1a columna es la num 0): '))
    return col #Nos devuelve el valor de la nueva columna o de la que ya estaba establecida en caso de que no la cambie

def columna_correcta_ia(grid, col):
    while grid[5][col] != 0:#Comprueba si la columna esta llena 
        col = randint(0, 6) #En caso que este llena la cambiara por otro valor aleatorio permitido
    return col#Nos devuelve el valor de la nueva columna o de la que ya estaba establecida en caso que no la cambie

def comprobar_valor(value):#Mira si el input introducido por el usuario es correcto
    while value > 6 or value < 0:
        value=int(input('Ha de ser un valor positivo entre 0 y 6'))
    return value#Nos devuelve el nuevo valor del usuario 

def comprobar_si_puede_ganar(grid, col):#Mira si la IA puede ganar
    for i in range(3):
        for j in range(7): #Revisa los casos posibles en vertical
            if grid[i][j] == 2 and grid[i + 1][j] == 2 and grid[i + 2][j] == 2 and grid[i+3][j] ==0:#Mira si hay 3 fichas del mismo color en vertical y la ultima posición está disponible
                col = j
                return col
    for i in range(6):
        for j in range(4): #Revisa los casos posibles en horizontal
            if grid[i][j] == 2 and grid[i][j + 1] == 2 and grid[i][j + 2] == 2 and grid[i][j+3] == 0:#Mira si hay 3 fichas del mismo color en horizontal y la posición que falta para  hacer 4 está disponible
                col = j + 3
                return col
            elif grid[i][j + 1] == 2 and grid[i][j + 2] == 2 and grid[i][j+3] == 2 and grid[i][j] == 0:#Mira si hay 3 fichas del mismo color en horizontal y la posición que falta para  hacer 4 está disponible
                col = j
                return col
            elif grid[i][j] == 2 and grid[i][j + 2] == 2 and grid[i][j + 3] == 2 and grid[i][j+1] == 0:#Mira si hay 3 fichas del mismo color en horizontal y la posición que falta para  hacer 4 está disponible
                col = j + 1
                return col
            elif grid[i][j] == 2 and grid[i][j + 1] == 2 and grid[i][j + 3] == 2 and grid[i][j+2] == 0:#Mira si hay 3 fichas del mismo color en horizontal y la posición que falta para  hacer 4 está disponible
                col = j + 2
                return col
    for i in range(3):
        for j in range(4):#Revisa los casos posibles en diagonal ascendiente
            if grid[i][j] == 2 and grid[i + 1][j + 1] == 2 and grid[i + 2][j + 2] == 2 and grid[i + 2][j + 3] == 2 and grid[i+3][j+3] == 0:#Mira si hay 3 fichas del mismo color en diagonal ascendiente y la posición que falta para hacer 4 está disponible, ademas que la de debajo a esta esté ocupada
                col = j + 3
                return col
            elif grid[i][j] == 2 and grid[i + 2][j + 2] == 2 and grid[i + 3][j + 3] == 2 and grid[i][j + 1] == 2 and grid[i+1][j+1] == 0:#Mira si hay 3 fichas del mismo color en diagonal ascendiente y la posición que falta para hacer 4 está disponible, ademas que la de debajo a esta esté ocupada
                col = j + 1
                return col
            elif grid[i][j] == 2 and grid[i + 1][j + 1] == 2 and grid[i + 3][j + 3] == 2 and grid[i + 1][j + 2] == 2 and grid[i+2][j+2] == 0:#Mira si hay 3 fichas del mismo color en diagonal ascendiente y la posición que falta para hacer 4 está disponible, ademas que la de debajo a esta esté ocupada
                col = j + 2
                return col
            elif grid[i + 1][j + 1] == 2 and grid[i+2][j+2] == 2 and grid[i + 3][j + 3] == 2 and grid[i][j] == 0:#Mira si hay 3 fichas del mismo color en diagonal ascendiente y la posición que falta para hacer 4 está disponible
                col = j
                return col
    for i in range(3, 6):
        for j in range(4):#Revisa todos los casos posibles en diagonal descendiente
            if grid[i][j] == 2 and grid[i - 1][j + 1] == 2 and grid[i - 2][j + 2] == 2 and grid[i - 4][j + 3] == 2 and grid[i-3][j+3] == 0:#Mira si hay 3 fichas del mismo color en diagonal descendiente y la posición que falta para hacer 4 está disponible
                col = j + 3
                return col
            elif grid[i][j] == 2 and grid[i - 1][j + 1] == 2 and grid[i - 3][j + 3] == 2 and grid[i - 3][j + 2] == 2 and grid[i-2][j+2] == 0:#Mira si hay 3 fichas del mismo color en diagonal descendiente y la posición que falta para hacer 4 está disponible
                col = j + 2
                return col
            elif grid[i][j] == 2 and grid[i - 2][j + 2] == 2 and grid[i - 3][j + 3] == 2 and grid[i - 2][j + 1] == 2 and grid[i-1][j+1] == 0:#Mira si hay 3 fichas del mismo color en diagonal descendiente y la posición que falta para hacer 4 está disponible
                col = j + 1
                return col
            elif grid[i-1][j+1] == 2 and grid[i - 2][j + 2] == 2 and grid[i - 3][j + 3] == 2 and grid[i][j] == 0:#Mira si hay 3 fichas del mismo color en diagonal descendiente y la posición que falta para hacer 4 está disponible
                col = j
                return col
    return col
def comprobar_si_puede_evitar_victoria(grid, col):#Misma funcion que la anterior pero esta mira que las posiciones sean del usuario, y en ese caso pondrá la pieza para evitar la victoria
    for i in range(3):
        for j in range(7):
            if grid[i][j] == 1 and grid[i + 1][j] == 1 and grid[i + 2][j] == 1 and grid[i+3][j] == 0:
                col = j
                return col
    for i in range(6):
        for j in range(4):
            if grid[i][j] == 1 and grid[i][j + 1] == 1 and grid[i][j + 2] == 1 and grid[i][j + 3] == 0:
                col = j + 3
                return col
            elif grid[i][j + 1] == 1 and grid[i][j + 2] == 1 and grid[i][j+3] == 1 and grid[i][j] == 0:
                col = j
                return col
            elif grid[i][j] == 1 and grid[i][j + 2] == 1 and grid[i][j + 3] == 1 and grid[i][j + 1] == 0:
                col = j + 1
                return col
            elif grid[i][j] == 1 and grid[i][j + 1] == 1 and grid[i][j + 3] == 1 and grid[i][j + 2] == 0:
                col = j + 2
                return col
                
    for i in range(3):
        for j in range(4):
            if grid[i][j] == 1 and grid[i + 1][j + 1] == 1 and grid[i + 2][j + 2] == 1 and grid[i + 2][j + 3] == 1 and grid[i+3][j+3] == 0:
                col = j + 3
                return col
            elif grid[i][j] == 1 and grid[i + 2][j + 2] == 1 and grid[i + 3][j + 3] == 1 and grid[i][j + 1] == 1 and grid[i+1][j+1] == 0:
                col = j + 1
                return col
            elif grid[i][j] == 1 and grid[i + 1][j + 1] == 1 and grid[i + 3][j + 3] == 1 and grid[i + 1][j + 2] == 1 and grid[i+2][j+2] == 0:
                col = j + 2
                return col
            elif grid[i + 1][j + 1] == 1 and grid[i+2][j+2] == 1 and grid[i + 3][j + 3] == 1 and grid[i][j] == 0:
                col = j
                return col
    for i in range(3, 6):
        for j in range(4):
            if grid[i][j] == 1 and grid[i - 1][j + 1] == 1 and grid[i - 2][j + 2] == 1 and grid[i - 4][j + 3] == 1 and grid[i-3][j+3] == 0:
                col = j + 3
                return col
            elif grid[i][j] == 1 and grid[i - 1][j + 1] == 1 and grid[i - 3][j + 3] == 1 and grid[i - 3][j + 2] == 1 and grid[i-2][j+2] == 0:
                col = j + 2
                return col
            elif grid[i][j] == 1 and grid[i - 2][j + 2] == 1 and grid[i - 3][j + 3] == 1 and grid[i - 2][j + 1] == 1 and grid[i-1][j+1] == 0:
                col = j + 1
                return col
            elif grid[i-1][j+1] == 1 and grid[i - 2][j + 2] == 1 and grid[i - 3][j + 3] == 1 and grid[i][j] == 0:
                col = j
                return col
    return col
```

```python
from ColabTurtle.Turtle import *
from random import randint
window_width = 700
window_height = 600
cursor_speed = 10
 
initializeTurtle(cursor_speed, (window_width, window_height))  # inicia la ventana
cell_side = 100
draw_board(cell_side)
 
tablero = [] #Generamos la matriz tablero
for i in range(6):
    tablero.append([0] * 7)
game_over = False 
turno = 0
print('Si quiere jugar 2 jugadores pulse 1', 'Si quiere jugar contra la IA pulse 2')
seleccion = (int(input('Que escoge? ')))
if seleccion == 1:
    while not game_over:
        if turno == 0:
            columna = int(input('En que columna desea colocar la pieza(Valor de 0-6)(JUGADOR 1): ')) #Pide al usuario la columna donde colocar la ficha
            comprobar_valor(columna) #Pasa por el filtro de que el valor sea uno correcto
            columna = columna_correcta(tablero, columna) #Comprueba que la columna no esté llena
            fila = fila_correcta(tablero, columna) #Mira en que fila ha de colocar la pieza
            place_piece(columna, fila, 'red', cell_side) #Coloca la pieza
            contador_pieza1(tablero, fila, columna) #Sustituye en la matriz la pieza para que sea contabilizada
            if ganador(tablero, 1):#Comprueba si hay ganador
                print('Ha ganado el jugador 1!')
                game_over = True
        else:#Mismo codigo pero para el jugador 2
            columna = int(input('En que columna desea colocar la pieza(Valor de 0-6)(JUGADOR 2): '))
            comprobar_valor(columna)
            columna = columna_correcta(tablero, columna)
            fila = fila_correcta(tablero, columna)
            place_piece(columna, fila, 'yellow', cell_side)
            contador_pieza2(tablero, fila, columna)
            if ganador(tablero, 2):
                print('Ha ganado el jugador 2!')
                game_over = True
        turno +=1#Contador del turno, irá sumando 1 a medida que pasen los turnos
        turno = turno % 2 #si el turno es par, al dividir quedará 0 y le tocará al jugador 1, si el turno es impar al dividir quedará 1 de resto y le tocará al jugador 2
elif seleccion == 2:
    while not game_over:
        if turno == 0:#Mismo codigo que en el 1vs1 para el jugador 1
            columna = int(input('En que columna desea colocar la pieza(Valor de 0-6)(JUGADOR 1): '))
            comprobar_valor(columna)
            columna = columna_correcta(tablero, columna)
            fila = fila_correcta(tablero, columna)
            place_piece(columna, fila, 'red', cell_side)
            contador_pieza1(tablero, fila, columna)
            if ganador(tablero, 1):
                print('Ha ganado el jugador 1!')
                game_over = True
        else:
            columna = randint(0, 6)#Primeramente hacemos que la columna de la IA sea un numero aleatorio
            columna = comprobar_si_puede_evitar_victoria(tablero, columna) #Comprueba si puede evitar victoria, en caso contrario, columna no cambiará el valor
            columna = comprobar_si_puede_ganar(tablero, columna) #Comprueba si puede ganar, en caso contrario, columna no cambiará el valor
            columna = columna_correcta_ia(tablero, columna) #Comprueba si la columna está llena, en caso contrario, columna no cambiará el valor
            fila = fila_correcta(tablero, columna) 
            place_piece(columna, fila, 'yellow', cell_side)
            contador_pieza2(tablero, fila, columna)
            if ganador(tablero, 2):
                print('Ha ganado la IA!')
                game_over = True
        turno += 1 #mismo contador que en el 1vs1
        turno = turno % 2
else:
  print('Escoja un valor correcto.') 
```

# **Informe de los algoritmos de mayor importancia**
**def ganador(grid, pieza):**

Queremos comprobar en cada movimiento si hay un ganador. Para esto definimos esta función:
Primeramente, comprobamos si obtenemos una victoria en vertical, usamos los dos for para que compruebe todas las posiciones posibles de la matriz donde pueda haber una linia ganadora, si las siguientes 3 posiciones en vertical son iguales y del mismo color, habrá un ganador. Hacemos lo mismo para las posibles victorias horizontales y diagonales tanto crecientes como decrecientes.

Definimos un contador para poder saber cuando hay una ficha colocada en cada sitio.

**def fila_correcta(grid, col):**

Queremos saber en que fila colocar la ficha. Para ello, primeramente intentamos que se coloque en la fila 0 de la columna que nos da el usuario, si esa posición es diferente de 0, es decir está ocupada, sumamos 1 a la fila hasta que esa posicion sea igual a 0, es decir que se pueda colocar ahí.

**def comprobar_valor(value):**

Comprobamos que el valor que nos da el usuario sea correcto, es decir entre 0 y 6, para ello abrimos un bucle si el valor no es correcto, donde pedirá al usuario otro valor hasta que de uno correcto. Nos pareció la mejor opción ya que así el estado actual de la partida se guardará hasta que se de un valor correcto, en caso contrario se perderia el progreso de la partida.

**def comprobar_si_puede_ganar(grid, col):**

Queremos hacer la máquina algo más inteligente; miramos una posible victoria en vertical, en este caso usamos el mismo mecanismo que para comprobar el ganador, es decir el rango de los for in es el numero de filas -3 en el caso de las victorias verticales, columnas -3 en el caso de las horizontales, filas -3 y columnas -3 en el caso de las diagonales crecientes y en el caso de las diagonales decrecientes, las filas cogeremos solo de la 3 a la 6 y las columnas todas - 3, esto es para evitar errores de indices, es decir que intente revisar una casilla que no está en el tablero. Ahora, lo hacemos sin igualar la ultima casilla, en este caso comprobamos que esta esté vacia, en ese caso colocamos ahi la ficha. En el caso de las horizontales, hay que tener en cuenta que puede ser que la ficha que falte no sea la ultima y sea uno del medio o la inicial, por eso hacemos elifs con los diferentes casos para que los pueda ver todos, para las diferentes diagonales hacemos el mismo procedimiento pero como está explicado en el código hay que tener en cuenta que la casilla de abajo esté llena para que si colocamos ahí la ficha obtengamos lo que queremos. En caso que no vea una posibilidad de ganar, devolvera el valor que ya teniamos de columna.

**def comprobar_si_puede_evitar_victoria(grid, col):**

La función en si es la misma, el algoritmo es igual solo que revisa que las fichas consecutivas sean del usuario y no de la máquina, estas dos funciones están separadas para poder dar más importancia a ganar que a evitar ganar, porque antes nos encontrabamos con que si lo haciamos todo en una función, la máquina daba prioridad a evitar una victoria del rival cuando la máquina podia ganar en ese mismo movimiento.
